```python
import pandas as pd
import pandera as pa
```

EXTRAÇÃO DOS DADOS

```python
df = pd.read_csv("ocorrencia.csv", parse_dates=['ocorrencia_dia'], dayfirst=True)
df.head()
```

```python
df.dtypes
```

**Make The Data Validation**

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String,pa.Check.str_length(2,2)),
        "ocorrencia_aerodromo":pa.Column(pa.String),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String,pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'),nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
        
        
    }
)
```

```python
schema.validate(df)
```

**Agrupar os dados com as ocorrencias da região Sudeste de 2012**

```python
filtro1=df.ocorrencia_dia.dt.year==2012
filtro2=df.ocorrencia_uf.isin(['SP','MG','ES','RJ'])
dfsudeste2010=df.loc[filtro1&filtro2]
dfsudeste2010
```

```python
dfsudeste2010.groupby(['ocorrencia_uf','ocorrencia_classificacao']).size().sort_values(ascending=False)
```

```python
filtro1=dfsudeste2010.total_recomendacoes > 0
filtro2 = dfsudeste2010.ocorrencia_cidade == 'SÃO PAULO'
dfsudeste2010.loc[filtro1&filtro2]
```

```python
filtro = dfsudeste2010.total_recomendacoes > 0
dfsudeste2010.loc[filtro].groupby(['ocorrencia_cidade']).total_recomendacoes.sum().sort_values()
```
